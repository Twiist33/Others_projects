### 4.3 Radar Charts

```python
# On importe la librarie pandas
import pandas as pd


# On stocke l'url souhaité et la table recherché
url2 = 'https://fbref.com/en/comps/Big5/stats/players/Big-5-European-Leagues-Stats'
df2 = pd.read_html(url2, attrs={"id": "stats_standard"})
```

```python
# On affiche un extrait de la table
df2 = df2[0]
df2.head(10)
```

```python
# On stocke cela dans un dataframe en enlevant les stats par 90 min
df2 = df2[df2.columns.drop(list(df2.filter(regex='Per 90')))]
```

```python
# On réaffiche les 1ères données de la table
df2.head()
```

```python
# On enlève la 1ère catégorie de colonnes
df2 = df2.droplevel(0,axis=1)
df2.head()
```

```python
# On va filtrer la liste de joueurs pour garder seulement les attaquants
df2 = df2[df2['Pos'].str.contains('FW')]
df2.head()
```

```python
# On va normaliser notre base de données de façon à garder seulement les joueurs ayant joué un nombre suffisant de minutes dans la saison
df2 = df2[df2['Min'].astype(int)>400]
df2.head()
```

```python
# On analyse les colonnes présentes dans la BDD
df2.columns
```

```python
# On transforme ces colonnes en flottant
df2[['90s','xG', 'xAG']] = df2[['90s','xG', 'xAG']].astype(float)
```

```python
# On transforme ces colonnes en entier
df2[['Gls','Ast', 'G+A','PrgC','PrgP', 'PrgR']] = df2[['Gls','Ast', 'G+A','PrgC','PrgP', 'PrgR']].astype(int)
```

```python
# On crée de nouvelles colonnes en effectuant des calculs par 90 minutes

df2['goals_per_90'] = df2['Gls'] / df2['90s']
df2['assists_per_90'] = df2['Ast'] / df2['90s']
df2['goals_assists_per_90'] = df2['G+A'] / df2['90s']
df2['progressive_carries_per_90'] = df2['PrgC'] / df2['90s']
df2['progressive_passes_per_90'] = df2['PrgP'] / df2['90s']
df2['progressive_receptions_per_90'] = df2['PrgR'] / df2['90s']
df2['expected_goals_per_90'] = df2['xG'] / df2['90s']
df2['expected_assists_per_90'] = df2['xAG'] / df2['90s']
```

```python
# On calcule ensuite le rang du joueur sur ces métriques
df2['Goals'] = (df2['goals_per_90'].rank(pct=True)*100).astype(int)
df2['Assists'] = (df2['assists_per_90'].rank(pct=True)*100).astype(int)
df2['Goals + Assists'] = (df2['goals_assists_per_90'].rank(pct=True)*100).astype(int)
df2['Progressive Carries'] = (df2['progressive_carries_per_90'].rank(pct=True)*100).astype(int)
df2['Progressive Passes'] = (df2['progressive_passes_per_90'].rank(pct=True)*100).astype(int)
df2['Progressive Receptions'] = (df2['progressive_receptions_per_90'].rank(pct=True)*100).astype(int)
df2['Expected Goals'] = (df2['expected_goals_per_90'].rank(pct=True)*100).astype(int)
df2['Expected Assists'] = (df2['expected_assists_per_90'].rank(pct=True)*100).astype(int)
```

```python
# On importe un radar
from mplsoccer import Radar
```

```python
# On recherche les informations du joueur Kylian Mbappe
mbappe = df2[df2['Player'] == 'Kylian Mbappé']
mbappe.head()
```

```python
# On affiche les colonnes disponibles
df2.columns
```

```python
# On selectionne les colonnes que l'on souhaite pour le radar
columns_to_plot = [
       'Goals', 'Assists', 'Goals + Assists',
       'Progressive Carries', 'Progressive Passes', 'Progressive Receptions',
       'Expected Goals', 'Expected Assists'
]

# On construit le radar
radar = Radar(
    params=columns_to_plot,
    min_range=[0 for _ in columns_to_plot],
    max_range=[100 for _ in columns_to_plot]
)
```

```python
# On construit le radar pour Mbappe
fig, ax = radar.setup_axis()

# On choisit le radar en cercle noire
rings_inner = radar.draw_circles(
    ax=ax, facecolor='white', edgecolor='black',lw=1, zorder=1
)

# On veut en valeur les résultats obtenus pour le joueur Kylian Mbappe
radar_output = radar.draw_radar(
    ax=ax,
    values= list(mbappe[columns_to_plot].values.flatten()),
    kwargs_radar={'facecolor': '#77BA99'},
    kwargs_rings={'facecolor': '#EFF0D1'}
)

# On choisit le texte pour afficher ces valeurs
range_labels = radar.draw_range_labels(
    ax=ax, fontsize= 15, fontproperties='monospace'
)

param_labels = radar.draw_param_labels(
    ax=ax, fontsize= 15, fontproperties='monospace'
)

# On rajoute du texte
ax.text(x=.5,y=1.02,s='Kylian Mbappe', fontsize=20, va='center', transform=ax.transAxes)
```

```python
# On va comparer avec Erling Haaland
haaland = df2[df2['Player'] == 'Erling Haaland']
haaland.head()
```

```python
# On construit un radar comparatif entre Mbappe et Haaland

fig, ax = radar.setup_axis()

# On choisit le radar en cercle noire
rings_inner = radar.draw_circles(
    ax=ax, facecolor='white', edgecolor='black',lw=1, zorder=1
)

# On veut en valeur les résultats obtenus pour le joueur Kylian Mbappe
radar_output = radar.draw_radar_compare(
    ax=ax,
    values= list(mbappe[columns_to_plot].values.flatten()),
    compare_values=list(haaland[columns_to_plot].values.flatten()),
    kwargs_radar={'facecolor': '#e63946', 'alpha': .5},
    kwargs_compare={'facecolor': '#669bbc', 'alpha': .5}
)

# On choisit le texte pour afficher ces valeurs
range_labels = radar.draw_range_labels(
    ax=ax, fontsize= 15, fontproperties='monospace'
)

param_labels = radar.draw_param_labels(
    ax=ax, fontsize= 15, fontproperties='monospace'
)
```

### 4.6 xG Flow Charts


```python
# On importe les données de tirs du match souhaité
import pandas as pd
url4 = 'https://fbref.com/en/matches/e81453d3/Liverpool-Real-Madrid-May-28-2022-Champions-League'
df5 = pd.read_html(url4,attrs = {"id":"shots_all"})[0]
```

```python
# On affiche les premières lignes de notre jeu de données
df5.head()
```

```python
# On enlève la 1ère colonne et les informations non souhaité

df5 = df5.drop(columns=[x for x in df5.columns if 'SCA' in x[0]])
df5.columns = df5.columns.droplevel(0)
```

```python
# On ré-affiche les premières lignes de notre jeu de données
df5.head()
```

```python
# On selectionne uniquement les variables souhaité
df5 = df5[['Minute','Player','Squad','xG','Outcome']]
```

```python
# On calcule le cumul des xG au cours du match par équipe
df5['cumulative_xG'] = df5.groupby('Squad')['xG'].cumsum()
```

```python
# On ré-affiche les lignes de notre jeu de données pour vérifier la nouvelle colonne
df5
```

```python
# On enlève les lignes vides
df5 = df5.dropna(subset=['Minute'])
```

```python
# On ajoute une colonne permettant de donner le numéro de la période de jeu au moment du tir
df5['half'] = df5['Minute'].apply(lambda x: 1 if int(x.split('+')[0]) <= 45 else 2)
```

```python
# On ré-affiche les lignes de notre jeu de données
df5
```

```python
# On additionne les minutes de jeu lorsque l'on entre dans le temps additionnel (ex : 90+2 devient 92)
df5['Minute'] = df5['Minute'].apply(lambda x:sum([int(y) for y in x.split('+')]))
```

```python
# On ré-affiche les lignes de notre jeu de données
df5
```

```python
# On construit notre graphique de xG cummulatifs sur le match par équipe
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,5))

# On associe les deux équipes à une couleur
liverpool_color = 'red'
real_madrid_color = 'blue'

for team in df5['Squad'].unique():
    team_df = df5[df5['Squad'] == team ]

    # On commence les lignes à 0 pour la 1ère période
    team_df = pd.concat([pd.DataFrame({'Squad': team, 'Minute': 0, 'xG' : 0, 'Outcome' : '', 'cumulative_xG' : 0, 'half' : 1},
                                      index=[0]),team_df])

    # On commence les lignes de la 2nd période selon les dernières valeurs de la 1ère période
    team_df = pd.concat([team_df[team_df['half'] == 1],
        pd.DataFrame({'Squad': team, 'Minute': 45, 'xG' : 0, 'Outcome' : '',
        'cumulative_xG' : team_df[team_df['half'] == 1]['cumulative_xG'].iloc[-1], 'half' : 2},
        index=[0]),team_df[team_df['half'] == 2]])
    
    # On écrit la boucle qui permettra d'afficher les lignes sur notre graphique
    for half in team_df['half'].unique():
        half_df = team_df[team_df['half'] == half]
        ax.plot(
            half_df['Minute'],
            half_df['cumulative_xG'],
            label = team,
            drawstyle = 'steps-post',
            c = liverpool_color if team=='Liverpool' else real_madrid_color
        )

# On affiche en légende les minutes suivantes
ax.set_xticks([0,45,90])

# On affiche les sections de chaque période
ax.text(22.5,-.25, 'First Half', ha='center', fontsize=12)
ax.text(67.5,-.25, 'Second Half', ha='center', fontsize=12)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# On écrit le titre et ses paramètres d'affichage

from highlight_text import fig_text

fig_text(
    .5,
    .95,
    '<Real Madrid> vs <Liverpool> - 2021/2022 UEFA Champions League Final',
    fontsize = 14,
    ha='center',
    va='center',
    ax=ax,
    highlight_textprops = [{'color': real_madrid_color}, {'color': liverpool_color}]
)
```

### 4.8 Pass Map

```python
# Importer les librairies nécéssaires

import pandas as pd
from statsbombpy import sb

# On charge les évenements d'un match
df6 = sb.events(match_id=3869685)
```

```python
# On affiche le décompte des évenements impliquant chaque joueur
print(df6.player.value_counts())
```

```python
# On filtre afin d'obtenir seulement les passes d'un joueur
df6 = df6[(df6['player'] == 'Aurélien Djani Tchouaméni') & (df6['type'] == 'Pass')]
```

```python
# On affiche les 1ère lignes de la table
df6.head()
```

```python
# On filtre les colonnes que l'on souhaite
df6 = df6[['location', 'pass_end_location', 'pass_outcome']]
```

```python
# On ré-affiche les 1ère lignes de la table
df6.head()
```

```python
# On va créer une colonne x et y pour séparer l'information de location

df6[['x','y']] = df6['location'].apply(pd.Series)
```

```python
# On ré-affiche les 1ère lignes de la table
df6.head()
```

```python
# On va créer une colonne end_x et end_y pour séparer l'information de end_pass_location et donne pour complété les passes ayant la mention NA

df6[['end_x','end_y']] = df6['pass_end_location'].apply(pd.Series)
df6['pass_outcome'] = df6['pass_outcome'].fillna("Complete")
```

```python
# On ré-affiche les 1ère lignes de la table
df6.head()
```

```python
# On informe dans une nouvelle colonne si la passe entre dans le dernier tiers du terrain

df6['entered_final_third'] = (df6['end_x'] >=80) & (df6['x'] <80)

# On décompte le nombre catégorisé ou non dans le dernier tiers
df6.entered_final_third.value_counts()
```

```python
# On rajoute la mention si la passe entre dans le dernier tiers du terrain
df6 = df6[df6['entered_final_third'] == True ]
```

```python
# On ré-affiche les 1ère lignes de la table
df6.head()
```

```python
# On va ensuite affiche les passes de Tchouameni sur le terrain

from mplsoccer import Pitch

pitch = Pitch(pitch_type='statsbomb')

fig, ax = pitch.draw()

# On stocke toutes les lignes du dataframe (passe entrant dans le dernier tiers de Tchouameni) dans une boucle for en informant sa réussite
for x in df6.to_dict(orient='records'):
    if x ['pass_outcome'] == 'Complete':
        pitch.lines(x['x'], x['y'], x['end_x'], x['end_y'], lw=5, transparent=True, comet=True, ax=ax, color='g')
    else:
        pitch.lines(x['x'], x['y'], x['end_x'], x['end_y'], lw=5, transparent=True, comet=True, ax=ax, color='r')

# On choisit le titre
ax.set_title("Aurélien Tchouameni Final Third Entries \n World Cup Final 2022", fontsize=12, fontfamily='monospace')

# On ajoute une légende
from matplotlib.lines import Line2D

legends_elements = [
    Line2D([0], [0], color='g', lw=4, label='Complete Pass'),
    Line2D([0], [0], color='r', lw=4, label='Incomplete Pass'),
]

# On affiche la légende

ax.legend(handles=legends_elements, loc='lower left')
```

### 4.9 Shot Map

```python
# On charge les données du match et les librairies
import pandas as pd
from statsbombpy import sb

df7 = sb.events(match_id=3869685)
```

```python
# On affiche les colonnes
df7.columns
```

```python
# On filtre les données pour garder seulement les tirs
df7 = df7[df7['type'] == 'Shot']

# Puis on garde seulement une liste de variables associées aux tirs

df7 = df7[['team', 'player', 'location', 'shot_outcome', 'shot_statsbomb_xg', 'shot_type']]
```

```python
# On affiche les 1ères lignes de la table filtrée
df7.head()
```

```python
# On filtrer pour exclure les pénaltys de notre shot map, n'étant pas représentatif d'une chance de but dans le jeu

df7 = df7[df7['shot_type'] != 'Penalty']
```

```python
# On ré-affiche les 1ères lignes de la table filtrée
df7.head()
```

```python
# On va créer une colonne x et y pour séparer l'information de location

df7[['x','y']] = df7['location'].apply(pd.Series)
```

```python
# On construit le terrain pour la shotmap

from mplsoccer import Pitch

pitch = Pitch(pitch_type='statsbomb')

fig, ax = pitch.draw()

# On sépare la shotmap en fonction de l'équipe de chaque coté du terrain et on met de couleur verte lorsqu'il s'agit d'un but
for x in df7.to_dict(orient='records'):
    if x['team'] == 'France':
        pitch.scatter(x=120-x['x'], y=80-x['y'], ax=ax, s=500*x['shot_statsbomb_xg'],
        ec='black', c='green' if x['shot_outcome'] == 'Goal' else 'white')
    else:
        pitch.scatter(x=x['x'], y=x['y'], ax=ax, s=500*x['shot_statsbomb_xg'],
        ec='black', c='green' if x['shot_outcome'] == 'Goal' else 'white')

# On écrit le titre
ax.set_title('2022 World Cup Final Shotmap', fontsize=20)

ax.text(30, 86, 'France', ha='center', fontsize=16)
ax.text(90, 86, 'Argentina', ha='center', fontsize=16)
```

```python
# Pour stocker les numéros des maillots
tactics = df9[~df9['tactics'].isnull()][['tactics', 'team']]

all_jersey_numbers = []
for y in tactics.to_dict(orient='records'):
    all_jersey_numbers.append(pd.DataFrame([{'player_id': x['player']['id'], 'jersey_number': x['jersey_number']} for x in y['tactics']['lineup']]))
    
jersey_numbers = pd.concat(all_jersey_numbers).drop_duplicates()
```

```python
# On calcule le temps de faàon à combiner les minutes et secondes du match dans une seule colonne
df9['newsecond'] = 60 * df9['minute'] + df9['second']

# On réordonne les données en fonction de cette colonnes
df9 = df9.sort_values(by=['newsecond'])
```

```python
# On affiche le résultat
df9['newsecond']
```

```python
# On crée deux nouvelles colonnes du passeurs et du receveur
df9['passer']= df9['player_id']
df9['recipient'] = df9['passer'].shift(-1)
```

```python
# On filtre afin de ne garder que les passes
passes_df = df9[df9['type']== 'Pass']
```

```python
# On modifie la colonne pass_outcome afin de déceler lorsque la passe est réussie
passes_df['pass_outcome'] = passes_df['pass_outcome'].fillna("Successful")
```

```python
# On vérifie

passes_df['pass_outcome']
```

```python
# On stocke dans une nouveau dataframe uniquement les passes réussies
completions = passes_df[passes_df['pass_outcome'] == 'Successful']
```

```python
# On cherche le moment où les 1ers changements français sont effectuées

sub_df = df9[df9['type'] == 'Substitution']
first_sub = sub_df['newsecond'].min()
```

```python
# On affiche la seconde de ce 1er changement
first_sub
```

```python
# Si il y a un changement avant la 1ère période, prendre comme point de référence le début de la seconde période
if first_sub  <= (60*45):
    first_sub = 60 * 45
```

```python
# On stocke les passes réussies avant le 1er changement
completions = completions[completions['newsecond'] < first_sub]
```

```python
# On calcule la position moyenne de chaque passeur français et on calcule son nombre
average_locations_and_counts = completions.groupby('passer').agg({'x': ['mean'], 'y': ['mean', 'count'] })
```

```python
# On renomme les colonnes
average_locations_and_counts.columns = ['x','y','count']
```

```python
# On vérifie
average_locations_and_counts.head()
```

```python
# On crée un nouveau dataframe pour cette fois-ci analyser le nombre de passes entre chaque joueur

passes_between = completions.groupby(['passer', 'recipient']).id.count().reset_index()
```

```python
# On renomme les colonnes du datframe
passes_between.rename({'id': 'pass_count'}, axis='columns', inplace=True)
```

```python
# On associe les deux dataframe
passes_between = passes_between.merge(average_locations_and_counts, left_on='passer', right_index=True)
passes_between = passes_between.merge(average_locations_and_counts, left_on='recipient', right_index=True, suffixes= ['','_end'])
```

```python
# On affiche notre table
passes_between.head()
```

```python
# On filtre le nombre de passes entre deux joueurs étant trop faible
passes_between = passes_between[passes_between['pass_count'] >= 4]
```

```python
# On utilise une fonction pour créer le réseau de passes avec les no

    
def pass_line_template(ax, x, y, end_x, end_y, line_color):
    ax.annotate(
        '',
        xy=(end_y,  end_x),
        xytext=(y, x),
        zorder=1,
        arrowprops=dict(arrowstyle='-|>', linewidth=4, color=line_color, alpha=.85)
    )
    
def pass_line_template_shrink(ax, x, y, end_x, end_y, line_color, dist_delta=1.2):
    dist = math.hypot(end_x - x, end_y - y)
    angle = math.atan2(end_y-y, end_x-x)
    upd_x = x + (dist - dist_delta) * math.cos(angle)
    upd_y = y + (dist - dist_delta) * math.sin(angle)
    pass_line_template(ax, x, y, upd_x, upd_y, line_color=line_color)
```
