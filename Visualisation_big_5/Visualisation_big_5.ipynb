### **Tableau Xg For Against / Big 5**

```python
# Import des librairies nécessaires
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lien de l'url que l'on va utiliser
url = 'https://fbref.com/en/comps/Big5/Big-5-European-Leagues-Stats'
```

```python
# Stockage dans un datframe de la table
df2 = pd.read_html(url, attrs={'id': 'big5_table'})[0]
```

```python
# Affichage des premières lignes du dataframe
df2.head()
```

```python
# Création d'un graphique du nombre de buts marqués en fonction des xG
fig, ax = plt.subplots()

#On construit un graphique de points avec les variables xG et Goals For
ax.scatter(df2['xG'], df2['GF'], alpha=.5)

# On choisit nos légendes et titre de graphique
ax.set_xlabel('Expected Goal')
ax.set_ylabel('Goal Scored')
ax.set_title('Goals vs Expected Goals in 2024/2025 European Big 5 Leagues')

# On choisit la taille de notre graphique en fonction de nos données
# On construit également la ligne qui représente la moyenne en fonction des deux valeurs
ax.plot([0,40],[0,40], color='black', linestyle='--')

# On calcule la différence entre le nombre de buts marquées et les xG marquées
df2['difference'] = df2['GF'] - df2['xG']

# On note les 5 équipes qui sur-performe et sous-performe
overperformers = df2.nlargest(5, 'difference')
underperformers = df2.nsmallest(5, 'difference')

for i in range(5):
    ax.text(overperformers.iloc[i]['xG'], overperformers.iloc[i]['GF'], overperformers.iloc[i]['Squad'])
    ax.text(underperformers.iloc[i]['xG'], underperformers.iloc[i]['GF'], underperformers.iloc[i]['Squad'])
```

### **Radar Charts / Big 5**

```python
# On importe la librarie pandas
import pandas as pd


# On stocke l'url souhaité et la table recherché
url2 = 'https://fbref.com/en/comps/Big5/stats/players/Big-5-European-Leagues-Stats'
df2 = pd.read_html(url2, attrs={"id": "stats_standard"})
```

```python
# On affiche un extrait de la table
df2 = df2[0]
df2.head(10)
```

```python
# On stocke cela dans un dataframe en enlevant les stats par 90 min
df2 = df2[df2.columns.drop(list(df2.filter(regex='Per 90')))]
```

```python
# On réaffiche les 1ères données de la table
df2.head()
```

```python
# On enlève la 1ère catégorie de colonnes
df2 = df2.droplevel(0,axis=1)
df2.head()
```

```python
# On va filtrer la liste de joueurs pour garder seulement les attaquants
df2 = df2[df2['Pos'].str.contains('FW')]
df2.head()
```

```python
# On va normaliser notre base de données de façon à garder seulement les joueurs ayant joué un nombre suffisant de minutes dans la saison
df2 = df2[df2['Min'].astype(int)>400]
df2.head()
```

```python
# On analyse les colonnes présentes dans la BDD
df2.columns
```

```python
# On transforme ces colonnes en flottant
df2[['90s','xG', 'xAG']] = df2[['90s','xG', 'xAG']].astype(float)
```

```python
# On transforme ces colonnes en entier
df2[['Gls','Ast', 'G+A','PrgC','PrgP', 'PrgR']] = df2[['Gls','Ast', 'G+A','PrgC','PrgP', 'PrgR']].astype(int)
```

```python
# On crée de nouvelles colonnes en effectuant des calculs par 90 minutes

df2['goals_per_90'] = df2['Gls'] / df2['90s']
df2['assists_per_90'] = df2['Ast'] / df2['90s']
df2['goals_assists_per_90'] = df2['G+A'] / df2['90s']
df2['progressive_carries_per_90'] = df2['PrgC'] / df2['90s']
df2['progressive_passes_per_90'] = df2['PrgP'] / df2['90s']
df2['progressive_receptions_per_90'] = df2['PrgR'] / df2['90s']
df2['expected_goals_per_90'] = df2['xG'] / df2['90s']
df2['expected_assists_per_90'] = df2['xAG'] / df2['90s']
```

```python
# On calcule ensuite le rang du joueur sur ces métriques
df2['Goals'] = (df2['goals_per_90'].rank(pct=True)*100).astype(int)
df2['Assists'] = (df2['assists_per_90'].rank(pct=True)*100).astype(int)
df2['Goals + Assists'] = (df2['goals_assists_per_90'].rank(pct=True)*100).astype(int)
df2['Progressive Carries'] = (df2['progressive_carries_per_90'].rank(pct=True)*100).astype(int)
df2['Progressive Passes'] = (df2['progressive_passes_per_90'].rank(pct=True)*100).astype(int)
df2['Progressive Receptions'] = (df2['progressive_receptions_per_90'].rank(pct=True)*100).astype(int)
df2['Expected Goals'] = (df2['expected_goals_per_90'].rank(pct=True)*100).astype(int)
df2['Expected Assists'] = (df2['expected_assists_per_90'].rank(pct=True)*100).astype(int)
```

```python
# On importe un radar
from mplsoccer import Radar
```

```python
# On recherche les informations du joueur Kylian Mbappe
mbappe = df2[df2['Player'] == 'Kylian Mbappé']
mbappe.head()
```

```python
# On affiche les colonnes disponibles
df2.columns
```

```python
# On selectionne les colonnes que l'on souhaite pour le radar
columns_to_plot = [
       'Goals', 'Assists', 'Goals + Assists',
       'Progressive Carries', 'Progressive Passes', 'Progressive Receptions',
       'Expected Goals', 'Expected Assists'
]

# On construit le radar
radar = Radar(
    params=columns_to_plot,
    min_range=[0 for _ in columns_to_plot],
    max_range=[100 for _ in columns_to_plot]
)
```

```python
# On construit le radar pour Mbappe
fig, ax = radar.setup_axis()

# On choisit le radar en cercle noire
rings_inner = radar.draw_circles(
    ax=ax, facecolor='white', edgecolor='black',lw=1, zorder=1
)

# On veut en valeur les résultats obtenus pour le joueur Kylian Mbappe
radar_output = radar.draw_radar(
    ax=ax,
    values= list(mbappe[columns_to_plot].values.flatten()),
    kwargs_radar={'facecolor': '#77BA99'},
    kwargs_rings={'facecolor': '#EFF0D1'}
)

# On choisit le texte pour afficher ces valeurs
range_labels = radar.draw_range_labels(
    ax=ax, fontsize= 15, fontproperties='monospace'
)

param_labels = radar.draw_param_labels(
    ax=ax, fontsize= 15, fontproperties='monospace'
)

# On rajoute du texte
ax.text(x=.3,y=1.02,s=' Radar Kylian Mbappe (Season 24/25)', fontsize=20, va='center', transform=ax.transAxes)
```

### **Pass Map**

```python
# Importer les librairies nécéssaires

import pandas as pd
from statsbombpy import sb

# On charge les évenements d'un match
df6 = sb.events(match_id=3869685)
```

```python
# On affiche le décompte des évenements impliquant chaque joueur
print(df6.player.value_counts())
```

```python
# On filtre afin d'obtenir seulement les passes d'un joueur
df6 = df6[(df6['player'] == 'Aurélien Djani Tchouaméni') & (df6['type'] == 'Pass')]
```

```python
# On affiche les 1ère lignes de la table
df6.head()
```

```python
# On filtre les colonnes que l'on souhaite
df6 = df6[['location', 'pass_end_location', 'pass_outcome']]
```

```python
# On ré-affiche les 1ère lignes de la table
df6.head()
```

```python
# On va créer une colonne x et y pour séparer l'information de location

df6[['x','y']] = df6['location'].apply(pd.Series)
```

```python
# On ré-affiche les 1ère lignes de la table
df6.head()
```

```python
# On va créer une colonne end_x et end_y pour séparer l'information de end_pass_location et donne pour complété les passes ayant la mention NA

df6[['end_x','end_y']] = df6['pass_end_location'].apply(pd.Series)
df6['pass_outcome'] = df6['pass_outcome'].fillna("Complete")
```

```python
# On ré-affiche les 1ère lignes de la table
df6.head()
```

```python
# On informe dans une nouvelle colonne si la passe entre dans le dernier tiers du terrain

df6['entered_final_third'] = (df6['end_x'] >=80) & (df6['x'] <80)

# On décompte le nombre catégorisé ou non dans le dernier tiers
df6.entered_final_third.value_counts()
```

```python
# On rajoute la mention si la passe entre dans le dernier tiers du terrain
df6 = df6[df6['entered_final_third'] == True ]
```

```python
# On ré-affiche les 1ère lignes de la table
df6.head()
```

```python
# On va ensuite affiche les passes de Tchouameni sur le terrain

from mplsoccer import Pitch

pitch = Pitch(pitch_type='statsbomb')

fig, ax = pitch.draw()

# On stocke toutes les lignes du dataframe (passe entrant dans le dernier tiers de Tchouameni) dans une boucle for en informant sa réussite
for x in df6.to_dict(orient='records'):
    if x ['pass_outcome'] == 'Complete':
        pitch.lines(x['x'], x['y'], x['end_x'], x['end_y'], lw=5, transparent=True, comet=True, ax=ax, color='g')
    else:
        pitch.lines(x['x'], x['y'], x['end_x'], x['end_y'], lw=5, transparent=True, comet=True, ax=ax, color='r')

# On choisit le titre
ax.set_title("Aurélien Tchouameni Final Third Entries \n World Cup Final 2022", fontsize=12, fontfamily='monospace')

# On ajoute une légende
from matplotlib.lines import Line2D

legends_elements = [
    Line2D([0], [0], color='g', lw=4, label='Complete Pass'),
    Line2D([0], [0], color='r', lw=4, label='Incomplete Pass'),
]

# On affiche la légende

ax.legend(handles=legends_elements, loc='lower left')
```

```python
# On enregistre notre graphique
fig.savefig('/Users/twiist/Documents/Doc/Projet/Other_mini_project/Visualisation_big_5/Pass_final_third_entries_Aurelien_Tchouameni.png', dpi=300, bbox_inches='tight')
```

### **Heatmaps**

```python
# On importe les librairies et le match

import pandas as pd
from statsbombpy import sb

df8 = sb.events(match_id=3869685)
```

```python
# On regarde les colonnes disponibles
df8.columns
```

```python
# On filtre les données pour n'avoir uniquement les passes

df8 = df8[df8['type'] == 'Pass']
```

```python
# On sépare la location en deux colonnes distinctes

df8[['x','y']] = df8['location'].apply(pd.Series)
```

```python
# On filtrer pour n'avoir uniquement les passes réussies
df8 = df8[df8['pass_outcome'].isna()]
```

```python
# On affiche cela sur un terrain

from mplsoccer import Pitch

pitch = Pitch(pitch_type='statsbomb', pitch_color='white', line_color='black', line_zorder=2)

fig, ax = pitch.draw()

df_fra = df8[df8['team_id']== 771]

kde = pitch.kdeplot(
    df_fra.x,
    df_fra.y,
    fill=True,
    thresh=0,
    n_levels=100,
    cut=10,
    ax=ax,
    cmap='plasma'
)

ax.set_title('France Starting Successfull Pass Locations vs Argentina WC 2022')
```
